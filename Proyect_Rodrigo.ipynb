Selected columns to be deleted later in the pipeline (they do not belong to the previous list)

```python
def get_columns_not_in_list(df, columns_to_keep):
    # Obtener todas las columnas del DataFrame
    all_columns = df.columns.tolist()

    # Obtener las columnas que no están en la lista de columnas a mantener
    columns_not_in_list = [col for col in all_columns if col not in columns_to_keep]

    return columns_not_in_list

# Ejemplo de uso:
columns_not_in_list = get_columns_not_in_list(alldata_prep, determined_columns)
print(columns_not_in_list)
```

## Transformation Pipelines

Drop unnecessary columns (the 'id' too, then we will add it again)

```python
from sklearn.base import BaseEstimator, TransformerMixin

# Definir un transformador personalizado
class EliminarColumnas(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        # Elimina las columnas originales categóricas
        X = X.drop(self.columns, axis=1)
        return X
```

Categorical -> Numerical

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import LabelEncoder

# Definir un transformador personalizado
class miLabelEncoder(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        self.le = LabelEncoder()
        # Concatenar todas las columnas categóricas antes de ajustar
        labels = X[self.columns].values.flatten()
        self.le.fit(labels)
        return self

    def transform(self, X):
        # Aplicar la transformación a todas las columnas categóricas
        return X[self.columns].apply(self.le.transform)
```

We can compute a 95% confidence interval for the test RMSE:

```python
from scipy import stats

confidence = 0.95
squared_errors = (final_predictions - y_test) ** 2
np.sqrt(stats.t.interval(confidence, len(squared_errors) - 1,
                         loc=squared_errors.mean(),
                         scale=stats.sem(squared_errors)))
```

We could compute the interval manually like this:

```python
m = len(squared_errors)
mean = squared_errors.mean()
tscore = stats.t.ppf((1 + confidence) / 2, df=m - 1)
tmargin = tscore * squared_errors.std(ddof=1) / np.sqrt(m)
np.sqrt(mean - tmargin), np.sqrt(mean + tmargin)
```

Alternatively, we could use a z-scores rather than t-scores:

```python
zscore = stats.norm.ppf((1 + confidence) / 2)
zmargin = zscore * squared_errors.std(ddof=1) / np.sqrt(m)
np.sqrt(mean - zmargin), np.sqrt(mean + zmargin)
```
